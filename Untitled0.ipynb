```python
import pandas as pd
import statsmodels.api as sm

# Create the data
data = {
    'TV': [230.1, 44.5, 17.2, 151.5, 180.8],
    'Radio': [37.8, 39.3, 45.9, 41.3, 10.8],
    'Social_Media': [69.2, 45.1, 69.3, 58.5, 58.4],
    'Sales': [22.1, 10.4, 9.3, 18.5, 12.9]
}
df = pd.DataFrame(data)

# Show correlation
print("Correlation with Sales:")
print(df.corr()['Sales'])

# TV has highest correlation, build model
X = df['TV']
y = df['Sales']
X_const = sm.add_constant(X)
model = sm.OLS(y, X_const).fit()

print("\n" + "="*50)
print("MODEL SUMMARY")
print("="*50)
print(model.summary())

print("\n" + "="*50)
print("INTERPRETATION")
print("="*50)
print(f"R-squared: {model.rsquared:.4f}")
print(f"Coefficient for TV: {model.params[1]:.4f}")
print(f"P-value: {model.pvalues[1]:.6f}")

if model.pvalues[1] < 0.05:
    print("✅ Statistically significant (p < 0.05)")
else:
    print("❌ Not statistically significant")

print("\n💡 RECOMMENDATION: Allocate budget to TV advertising")
```
